### Dataset

```python
def select_data_set(diretorio):

    imagens = list()
    labels = list()

    for subdir in listdir(diretorio):
        # path
        path = diretorio + subdir + '/'

        if not isdir(path):
            continue
        imagens, labels = load_classes(path, subdir, imagens, labels)

    return imagens, labels
```

```python
def select_image(filename):
    # load image from file
    image = Image.open(filename)
    # convert to RGB, if needed
    image = image.convert('RGB')
    image = image.resize((150,150))
    # convert to array
    return np.asarray(image)
```

```python
def load_classes(diretorio, classe, imagens, labels):
    # iterando arquivos

    for filename in listdir(diretorio):

        path = diretorio + filename
        
        try:
            imagens.append(select_image(path))
            labels.append(classe)
        except:
            print("Erro ao ler imagem {}".format(path))

    return imagens, labels
```

```python
def create_adversarial_pattern(input_image, input_label):
  with tf.GradientTape() as tape:
    tape.watch(input_image)
    prediction = model_covid(input_image)
    loss = loss_object(input_label, prediction)

  # Get the gradients of the loss w.r.t to the input image.
  gradient = tape.gradient(loss, input_image)
  # Get the sign of the gradients to create the perturbation
  signed_grad = tf.sign(gradient)
  return signed_grad
```

### Metricas para definir a qualidade das imagens

```python
def PSNR(original, noise):
    _,m,n,c = original.shape
    mse = (1/m*n)*np.sum((noise - original)**2)
    L = np.max(original) 
    psnr = 10* np.log10((L**2/mse))

    return psnr
```

```python
def VPSNR(original, noise):
    _,m,n,c = original.shape
    mse = (1/m*n)*np.sum((noise - original)**2)
    sigma_x = np.std(original)
    singma_y = np.std(noise)
    vmse = mse/(1 + 0.5 * np.sqrt(sigma_x*singma_y))
    L = np.max(original) 
    psnr = 10* np.log10((L**2/vmse))
    
    return psnr
```

### Definindo as images aleatorias

```python
def generate_random_samples(no_images, no_epsilon, images):
    epsilon = np.random.random_sample(no_epsilon)
    #selecionar os index das imagens
    images_index = np.random.choice(np.arange(0,images.shape[0]-1, dtype=np.int), no_images)
    
    return images_index, epsilon
```

### Para executar o ataque

```python
def run_attack(image, model, adv_model, epsilon):
    #aplica o ataque
    #img = tf.convert_to_tensor(image.reshape(1,150,150,3))
    img = image.reshape(1,224,224,3)
    adv_sample_aux = fast_gradient_method(logits_model, img, epsilon, np.inf, targeted=False)
    #faz a predição da imagem atacada
    pred_adv = model.predict(adv_sample_aux)
    label_adv =  np.argmax(pred_adv[0]);
    #faz a predicao da image original
    pred_orig = model.predict(img)
    label_orig = np.argmax(pred_orig[0])
    #calcula a probabilidade da image pertencer a classe predita
    # pred_prob = model.predict(img, verbose=1)
    # pred_adv_prob = model.predict(adv_sample_aux, verbose=1)
    # probl = np.max(pred_prob[0])
    # probl_adv = np.max(pred_adv_prob[0])
    #calcula as metricas
    vpsrn = tf.image.psnr(img, adv_sample_aux, max_val=1.0)
  
    #return label da image original, o label da image atacada e a metrica de qualidade entre as duas
    return label_orig, label_adv, vpsrn
```

```python
def run_experiments(images_batch, label_batch, model, adv_model, no_epsilon=30, no_images=10):
    imgs, eps = generate_random_samples(no_images, no_epsilon, images_batch)
    labels_true, labels_adv, labels_orig, epsilons, qualities, img_idx = [],[],[],[],[],[]
    for idx in imgs:
        img = images_batch[idx]
        for e in eps:
            labels_true.append(label_batch[idx])
            label_orig, label_adv, vpsrn = run_attack(img, model, adv_model, e)
            labels_adv.append(label_adv)
            labels_orig.append(label_orig)
            epsilons.append(e)
            qualities.append(vpsrn)
            img_idx.append(idx)
  
    return labels_true, img_idx, labels_adv, labels_orig, epsilons, qualities
```

### Run tests

```python
data = pd.DataFrame(columns=["index","lab_true", "lab_adv", "lab_orig", "epsilon", "quality"])
logits_model = tf.keras.Model(model_covid.input,model_covid.layers[-1].output)
data["lab_true"], data["index"], data["lab_adv"], data["lab_orig"], data["epsilon"], data["quality"] = run_experiments(images, lables, model_covid, logits_model, 2, 100)
```

```python
data_best = data[data['lab_adv'] != data["lab_true"]]
```

```python
def calculate_metrics(y_true, y_adv, y_pred, y_score_adv, y_score_pred):
    pred_metrics = {'accuracy':0.0, 'recall': 0.0, 'f1-score': 0.0, 'auc': 0.0}
    adv_metrics = {'accuracy':0.0, 'recall': 0.0, 'f1-score': 0.0, 'auc': 0.0}
    #calcula a acuracia
    adv_metrics['accuracy'] = metrics.accuracy_score(y_true, y_adv)
    pred_metrics['accuracy'] = metrics.accuracy_score(y_true, y_pred)
    #calcula recall
    adv_metrics['recall'] = metrics.recall_score(y_true, y_adv, average='weighted')
    pred_metrics['recall'] = metrics.recall_score(y_true, y_pred, average='weighted')
    #calcula f1-score
    adv_metrics['f1-score'] = metrics.f1_score(y_true, y_adv, average='weighted')
    pred_metrics['f1-score'] = metrics.f1_score(y_true, y_pred, average='weighted')
    #calcula auc
    #adv_metrics['auc'] = metrics.roc_auc_score(y_true, y_score_adv, multi_class='ovo')
    #pred_metrics['auc'] = metrics.roc_auc_score(y_true, y_score_pred, multi_class='ovo')

    return pred_metrics, adv_metrics
```
