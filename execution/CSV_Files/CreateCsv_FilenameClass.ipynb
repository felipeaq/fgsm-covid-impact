```python
import pandas as pd
import os
import numpy as np
```

```python
# Cria uma lista com o os filenames das imagens contidas na pasta
#lista1 = os.listdir('/home/karem/1_Datasets/FullDataset_covid/Original/padchest/covid')
#lista2 = os.listdir('/home/karem/1_Datasets/FullDataset_covid/Original/padchest/nonCovid/')

lista1 = os.listdir('/home/karem/DisciplinaAgma/Rede_ArtigoSPIE/Covid19_RadiograpyDataset/COVID/')
lista2 = os.listdir('/home/karem/DisciplinaAgma/Rede_ArtigoSPIE/Covid19_RadiograpyDataset/Normal/')
#lista2
```

```python
# Renomeia a lista - tira a extensão
import re
new_lista1 = []
for item in lista1:
    #item = 'COVID/' + item # add o nome da pasta para facilitar no caminho de carregamento do arquivo
    l1 = re.sub('.png', '', str(item))
    #bar = re.sub('\)', '', bar)
    new_lista1.append(l1)

       

new_lista2 = []
for item in lista2:
    #item = 'Normal/' + item
    l2 = re.sub('.png', '', str(item))
    #bar = re.sub('\)', '', bar)
    new_lista2.append(l2)

#new_lista1
```

```python
# Converte a lista em Df
#df1 = pd.DataFrame()
#df1['folder'] = 'COVID/'
#df1.loc[:, 'folder'] = 'COVID/'

df1 = pd.DataFrame(new_lista1,columns=['image_name'])
df1.insert(0, 'folder', 'COVID/')


df2 = pd.DataFrame(new_lista2,columns=['image_name'])
df2.insert(0, 'folder', 'Normal/')
df2
```

```python
# Junta os Df
df_concat = pd.concat([df1, df2])
df_concat
```

```python
# 80% dos dados são utilizados para treino e validação e 20% para teste
def train_validate_test_split(df, train_percent=.7, validate_percent=.1, seed=None):    
    np.random.seed(seed)
    perm = np.random.permutation(df.index)    
    m = len(df.index)    
    train_end = int(train_percent * m)    
    validate_end = int(validate_percent * m) + train_end
    train = df.iloc[perm[:train_end]]
    validate = df.iloc[perm[train_end:validate_end]]
    test = df.iloc[perm[validate_end:]]
    return train, validate, test
```

```python
from sklearn.utils import shuffle
df_concat = shuffle(df_concat)
df_concat = shuffle(df_concat)
df_concat = shuffle(df_concat)
```

```python
# Divide o conjunto em treino e teste
train, validate, test = train_validate_test_split(df=df_concat)
print(len(train), len(validate), len(test))
```
